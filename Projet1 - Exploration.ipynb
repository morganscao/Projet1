```python
# Comparaison avec le grade
Xscorefr = dfScore.nutrition_score_fr_100g
Xscoreuk = dfScore.nutrition_score_uk_100g

def GraphWithSize(XX, yy, titre):
    fig = plt.figure(figsize=(8, 4))
    ax = plt.subplot()
    sizes = {} # clé : coordonnées ; valeur : nombre de points à ces coordonnées
    for (yt, yp) in zip(list(XX), list(yy)):
        if (yt, yp) in sizes:
            sizes[(yt, yp)] += 1
        else:
            sizes[(yt, yp)] = 1
    keys = sizes.keys()
    ax.scatter([k[0] for k in keys], # score en abscisse
    [k[1] for k in keys], # grade en ordonnée
    s=[(sizes[k])/100 for k in keys]) # taille du marqueur

    ax.set_title(titre)
    
GraphWithSize(Xscorefr, yGrade, "score fr vs grade")
GraphWithSize(Xscoreuk, yGrade, "score uk vs grade")
GraphWithSize(Xscorefr, Xscoreuk, "score fr vs score uk")
```

### Groupes

```python
# On enlève les grades nulls
dfGroup = dataraw[pd.notnull(dataraw.nutrition_grade_fr)]
colGrade = 'nutrition_grade_fr'

def PieCategorie(p_df, p_col):
    print("Répartition par grade des éléments de", p_col)
    dftmp2 = p_df[pd.notnull(p_df[p_col])][p_df[p_col] != '0']
    dftmp2 = dftmp2[dftmp2[p_col] != 'unknown']
    # Parcours des différentes valeurs de la catégorie
    for grp in dftmp2[p_col].unique():
        # Réduction de la matrice
        dftmp = dftmp2[[p_col, colGrade]][dftmp2[p_col] == grp]
        nb = dftmp.shape[0]
        dftmp = dftmp.groupby([p_col, colGrade])
        dftmp = dftmp.size().reset_index(name='counts')
        # Affichage
        fig = plt.figure(figsize=(5, 5))
        ax = fig.add_subplot(111)
        ax.pie(dftmp.counts, labels=dftmp.nutrition_grade_fr)
        ax.set_title(grp + " : " + (str)(nb))
        plt.legend()
        
PieCategorie(dfGroup, 'pnns_groups_1')
```

## Analyse multivariée

### Corrélations

```python
corr1 = dataraw.corr()
corr1
```

```python
# Masquage de la diagonale
mask = np.ones(corr1.columns.size) - np.eye(corr1.columns.size)
corrMasqued = mask * corr1

fig, ax = plt.subplots(figsize=(16,10))
ax.set_title('Corrélation des valeurs nutritionnelles')
heatmap = ax.pcolor(corrMasqued, cmap=plt.cm.Greens)
```

```python
# Analyse de la matrice de corrélation - Colonnes corrélées
corrCols = []
for col in corrMasqued.columns.values:
    # Si la feature a déjà été traitée on passe
    if np.in1d([col], corrCols):
        continue
    
    # Récupération des features fortement corrélées 
    corr = corrMasqued[abs(corrMasqued[col]) > 0.5].index
    corrCols = np.union1d(corrCols, corr)
    
    if corr.shape[0] > 0:
        print("%s corrélé à : %s" % (col, corr.tolist()))
```

```python
# Analyse de la matrice de corrélation - Colonnes non corrélées au score
col = 'nutrition_score_fr_100g'
nonCorrCols = []

# Récupération des features non corrélées 
corr = corr1[abs(corr1[col]) < 0.1].index
nonCorrCols = np.union1d(nonCorrCols, corr)

if corr.shape[0] > 0:
    print("%s non corrélé à : %s" % (col, corr.tolist()))
```

```python
def RemoveNonCorrColumns(p_df, p_drops):
    ret = p_df.copy()
    # Supprimons les colonnes sans lien avec le score
    for col in p_drops:
        ret = DropCol(ret, col)
    print(ret.shape)
    print("%i colonnes en moins" % (p_df.shape[1]-ret.shape[1]))
    return ret
```

```python
my_data = RemoveNonCorrColumns(dataraw, nonCorrCols)
```

```python
corr1 = my_data.corr()
mask = np.ones(corr1.columns.size) - np.eye(corr1.columns.size)
corrMasqued = mask * corr1
fig, ax = plt.subplots(figsize=(16,10))
ax.set_title('Corrélation des valeurs nutritionnelles')
heatmap = ax.pcolor(corrMasqued, cmap=plt.cm.Greens)
```

### Seaborn

```python
import seaborn as sns
def GraphSeaborn(p_df, p_cat, p_continue):
    cols = []
    count = 0
    p_df = p_df[pd.notnull(p_df[p_cat])]
    for i in p_df.columns:
        if p_continue in i:
            continue
        count += 1
        cols.append(i)
        if count > 0 and count % 4 == 0:
            cols.append(p_cat)
            #print (cols)
            
            # Si toutes les valeurs nutritives sont nulles on enlève la ligne
            d = p_df[sum(p_df[cols], axis=1) > 0]
            print(d.shape)

            sns.pairplot(d[cols], hue=p_cat, size=2.5);
            cols.clear()

    if len(cols) > 0:
        cols.append(p_cat)
        #print (cols)
        sns.pairplot(p_df[cols], hue=p_cat, size=2.5);
```

```python
dfreg = DropCol(my_data, 'pnns_groups_1')
dfreg = DropCol(dfreg, 'pnns_groups_2')
GraphSeaborn(dfreg, 'nutrition_grade_fr', 'nutrition')
```

### Scattering

On retrouve dans ces analyses les ingrédients dont il ne faut pas abuser (énergie, graisses saturées, sucre, sodium...), mais on voit que le score n'est pas corrélé linéairement, la formule est un peu plus complexe. 


```python
# Analyse des valeurs positives
def GraphScatter(p_df, p_score, p_continue):
    print("\tAffichage des features en fonction du score")
    count = 0
    for i in p_df.columns:
        if p_continue in i:
            continue
        dflight = p_df[p_df[i] > 0]
        mod = count%12
        if mod == 0:
            fig = plt.figure(figsize=(16, 12))
        ax = fig.add_subplot(3,4, (mod+1))
        ax.scatter(dflight[i], dflight[p_score], s=1, color='blue')
        ax.set_title(i, fontsize=14)
        count += 1
        
dfreg = DropCol(my_data, 'pnns_groups_1')
dfreg = DropCol(dfreg, 'pnns_groups_2')
GraphScatter(dfreg, 'nutrition_score_fr_100g', 'nutrition')
```

### PCA

```python
my_data = RemoveNonCorrColumns(dataraw, nonCorrCols)
#my_data = dataraw

dfPCA = DropCol(my_data, 'pnns_groups_1')
dfPCA = DropCol(dfPCA, 'pnns_groups_2')
dfPCA = DropCol(dfPCA, 'nutrition_grade_fr')
dfPCA = DropCol(dfPCA, 'nutrition_score_fr_100g')

X = dfPCA.values
y = my_data.nutrition_score_fr_100g

from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
from sklearn.decomposition import PCA

# Minimum percentage of variance we want to be described by the resulting transformed components
variance_pct = .8

# Create PCA object
pca = PCA(n_components=variance_pct)

# Transform the initial features
X_transformed = pca.fit_transform(X_scaled, y)
pcaDataFrame = pd.DataFrame(X_transformed)

print (pcaDataFrame.shape[1], "features sur", X_scaled.shape[1], "décrivent", str(variance_pct)[1:], "% de la variance")
```
