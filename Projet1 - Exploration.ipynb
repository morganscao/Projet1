```python
# Comparaison avec le grade
Xscorefr = df.nutrition_score_fr_100g
Xscoreuk = df.nutrition_score_uk_100g

def GraphWithSize(XX, yy, titre):
    fig = plt.figure(figsize=(8, 4))
    ax = plt.subplot()
    sizes = {} # cl√© : coordonn√©es ; valeur : nombre de points √† ces coordonn√©es
    for (yt, yp) in zip(list(XX), list(yy)):
        if (yt, yp) in sizes:
            sizes[(yt, yp)] += 1
        else:
            sizes[(yt, yp)] = 1
    keys = sizes.keys()
    ax.scatter([k[0] for k in keys], # score en abscisse
    [k[1] for k in keys], # grade en ordonn√©e
    s=[(sizes[k])/100 for k in keys]) # taille du marqueur

    ax.set_title(titre)
    
GraphWithSize(Xscorefr, yGrade, "score fr vs grade")
GraphWithSize(Xscoreuk, yGrade, "score uk vs grade")
GraphWithSize(Xscorefr, Xscoreuk, "score vs score")
```

## Analyse multi vari√©e - Corr√©lations

```python
corr1 = dataraw.corr()
mask = np.ones(corr1.columns.size) - np.eye(corr1.columns.size)
corr1 = mask * corr1
corr1
```

```python
fig, ax = plt.subplots(figsize=(16,10))
ax.set_title('Corr√©lation des valeurs nutritionnelles')
heatmap = ax.pcolor(corr1, cmap=plt.cm.Greens)
```

```python
# Analyse de la matrice de corr√©lation
drops = []
for col in corr1.columns.values:
    # Si la feature a d√©j√† √©t√© trait√©e on passe
    if np.in1d([col],drops):
        continue
    
    # R√©cup√©ration des features fortement corr√©l√©es 
    corr = corr1[abs(corr1[col]) > 0.5].index
    drops = np.union1d(drops, corr)
    
    if corr.shape[0] > 0:
        print("%s est fortement corr√©l√©e √† : %s" % (col, corr.tolist()))

#corr1
```

```python
# R√©sultats des analyses sur diff√©rents √©chantillons : 

#energy_100g est fortement corr√©l√©e √† : ['fat_100g', 'saturated_fat_100g', 'carbohydrates_100g', 'nutrition_score_fr_100g']
#salt_100g est fortement corr√©l√©e √† : ['sodium_100g']
#vitamin_a_100g est fortement corr√©l√©e √† : ['vitamin_c_100g']
#vitamin_b1_100g est fortement corr√©l√©e √† : ['vitamin_b6_100g']
#vitamin_b1_100g est fortement corr√©l√©e √† : ['vitamin_b2_100g']
#vitamin_b2_100g est fortement corr√©l√©e √† : ['vitamin_pp_100g', 'vitamin_b12_100g']
#monounsaturated_fat_100g est fortement corr√©l√©e √† : ['polyunsaturated_fat_100g']
#sugars_100g est fortement corr√©l√©e √† : ['carbohydrates_100g', 'nutrition_score_fr_100g']


# Regroupement des colonnes corr√©l√©es
my_data = dataraw.copy()

try:
    my_data['energie_et_fat_100g'] = my_data.energy_100g + my_data.fat_100g + my_data.saturated_fat_100g + my_data.carbohydrates_100g
    my_data = my_data.drop(['energy_100g', 'fat_100g', 'saturated_fat_100g', 'carbohydrates_100g'], axis=1)
except:
    print("Suppressions d√©j√† faites")

try:
    my_data['salt_et_sodium_100g'] = my_data.salt_100g + my_data.sodium_100g
    my_data = my_data.drop(['salt_100g', 'sodium_100g'], axis=1)
except:
    print("Suppressions d√©j√† faites")

try:
    my_data['saturated_100g'] = my_data.monounsaturated_fat_100g + my_data.polyunsaturated_fat_100g
    my_data = my_data.drop(['monounsaturated_fat_100g', 'polyunsaturated_fat_100g'], axis=1)
except:
    print("Suppressions d√©j√† faites")

try:
    my_data['vitA_et_vitC_100g'] = my_data.vitamin_a_100g + my_data.vitamin_c_100g
    my_data = my_data.drop(['vitamin_a_100g', 'vitamin_c_100g'], axis=1)
except:
    print("Suppressions d√©j√† faites")

try:
    my_data['vitB_et_vitPP_100g'] = my_data.vitamin_b1_100g + my_data.vitamin_b2_100g + my_data.vitamin_b6_100g + my_data.vitamin_b12_100g + my_data.vitamin_pp_100g
    my_data = my_data.drop(['vitamin_b1_100g', 'vitamin_b2_100g', 'vitamin_b12_100g', 'vitamin_b6_100g', 'vitamin_pp_100g'], axis=1)
except:
    print("Suppressions d√©j√† faites")
    
my_data.head()
```

### Seaborn

```python
import seaborn as sns
def GraphSeaborn(p_df, p_cat, p_continue):
    cols = []
    count = 0
    p_df = p_df[pd.notnull(p_df[p_cat])]
    for i in p_df.columns:
        if p_continue in i:
            continue
        count += 1
        cols.append(i)
        if count > 0 and count % 4 == 0:
            cols.append(p_cat)
            #print (cols)
            
            # Si toutes les valeurs nutritives sont nulles on enl√®ve la ligne
            d = p_df[sum(p_df[cols], axis=1) > 0]
            print(d.shape)

            sns.pairplot(d[cols], hue=p_cat, size=2.5);
            cols.clear()

    if len(cols) > 0:
        cols.append(p_cat)
        #print (cols)
        sns.pairplot(p_df[cols], hue=p_cat, size=2.5);
```

```python
GraphSeaborn(my_data, 'nutrition_grade_fr', 'nutrition')
```

### Scattering

On retrouve dans ces analyses les ingr√©dients dont il ne faut pas abuser (√©nergie, graisses satur√©es, sucre, sodium...), mais on voit que le score n'est pas corr√©l√© lin√©airement, la formule est un peu plus complexe. 


```python
# Analyse des valeurs positives
def GraphScatter(p_df, p_score, p_continue):
    count = 0
    for i in p_df.columns:
        if p_continue in i:
            continue
        dflight = p_df[p_df[i] > 0]
        mod = count%12
        if mod == 0:
            fig = plt.figure(figsize=(16, 12))
        ax = fig.add_subplot(3,4, (mod+1))
        ax.scatter(dflight[i], dflight[p_score], s=5, color='blue')
        ax.set_title(i, fontsize=14)
        count += 1
        
GraphScatter(my_data, 'nutrition_score_fr_100g', 'nutrition')
```

### ANOVA

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
model = ols('nutrition_score_fr_100g ~ fat_100g + saturated_fat_100g', data=dataraw).fit()
#print(model.summary())
table1 = anova_lm(model)
#print(table1)
if table1['PR(>F)']['fat_100g'] < 0.05:
    print('fat_100g', 'est tr√®s corr√©l√© au score')
```

```python
df = dataraw[pd.notnull(dataraw.nutrition_score_fr_100g)]
print(df.shape)
for i in dataraw.columns:
    if 'nutrition' in i:
        continue
    formule = 'nutrition_score_fr_100g ~ ' + i
    model = ols(formule, data=dataraw).fit()
    #print(model.summary())
    table1 = anova_lm(model)
    #print(table1)
    if table1['PR(>F)'][i] < 0.05:
        print(i, 'est tr√®s corr√©l√© au score, test de Fisher =', table1['PR(>F)'][i])
    else:
        print('\t', i, 'PAS tr√®s corr√©l√© au score, test de Fisher =', table1['PR(>F)'][i])
```

Avec l'ANOVA tout semble √™tre corr√©l√© au score

### PCA

```python
# transformer les donn√©es en array numpy
dfPCA = my_data.drop('nutrition_score_fr_100g', axis=1)
if 'nutrition_grade_fr' in my_data.columns:
    dfPCA = dfPCA.drop(['nutrition_grade_fr'], axis=1)

X = dfPCA.values
y = my_data.nutrition_score_fr_100g

from sklearn import preprocessing
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
```

```python
from sklearn.decomposition import PCA

# Minimum percentage of variance we want to be described by the resulting transformed components
variance_pct = .80

# Create PCA object
pca = PCA(n_components=variance_pct)

# Transform the initial features
X_transformed = pca.fit_transform(X_scaled, y)
pcaDataFrame = pd.DataFrame(X_transformed)

print (pcaDataFrame.shape[1], " components describe ", str(variance_pct)[1:], "% of the variance")
```
